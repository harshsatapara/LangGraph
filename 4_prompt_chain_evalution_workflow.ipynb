```python
from langgraph.graph import StateGraph, START, END
from langchain_community.chat_models.oci_generative_ai import ChatOCIGenAI
from typing import TypedDict
from dotenv import load_dotenv 
from langchain_openai import ChatOpenAI
```

```python
load_dotenv()

model = ChatOpenAI()
```

```python
class BlogEvalState(TypedDict):
    title:str
    outline:str
    content: str
    evalution: str
```

```python
def BlogOutline(state:BlogEvalState) -> BlogEvalState:
    title = state['title']
    prompt = f'Generate a details outline for a blog on the topic -{title}'
    outline = model.invoke(prompt).content
    state['outline'] = outline
    return state

def GenBlog(state:BlogEvalState) -> BlogEvalState:
    title = state['title']
    outline = state['outline']
    prompt = f'Write detailed blog on the title - {title} using the following outline \n {outline}'
    blog = model.invoke(prompt).content
    state['content'] = blog
    return state

def BlogEvalute(state: BlogEvalState) -> BlogEvalState:
    outline = state['outline']
    blog = state['content']
    prompt = f'Evaluate generated blog from provided out line.Provide rating out of 10 points. \n Outline: {outline} \n Blog: {blog}'
    Beval = model.invoke(prompt).content
    state['evalution'] = Beval
    return state
```

```python
graph = StateGraph(BlogEvalState)

graph.add_node("BlogOutline",BlogOutline)
graph.add_node("GenBlog",GenBlog)
graph.add_node("BlogEvalute",BlogEvalute)

graph.add_edge(START,"BlogOutline")
graph.add_edge("BlogOutline","GenBlog")
graph.add_edge("GenBlog","BlogEvalute")
graph.add_edge("BlogEvalute",END)

workflow = graph.compile()

initial_state = {"title": "Importance of education in our life."}
final_state = workflow.invoke(initial_state)
print(final_state)
```
