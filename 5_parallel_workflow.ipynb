```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class calState(TypedDict):
    num1: int
    num2: int
    sum: int
    sub: int
    mul: int
    div: float
```

```python
def getSum(state: calState):
    add = state.get("num1") + state.get("num2")
    return {"sum": add}

def getSub(state: calState):
    sub = state.get("num1") - state.get("num2")
    return {"sub":sub}

def getMul(state: calState):
    mul = state.get("num1") * state.get("num2")
    return {"mul": mul}

def getDiv(state: calState):
    div = state.get("num1") / state.get("num2")
    return {"div" : div}
```

```python
graph = StateGraph(calState)

graph.add_node("getSum",getSum)
graph.add_node("getSub",getSub)
graph.add_node("getMul",getMul)
graph.add_node("getDiv",getDiv)

graph.add_edge(START,"getSum")
graph.add_edge(START,"getSub")
graph.add_edge(START,"getMul")
graph.add_edge(START,"getDiv")

graph.add_edge("getSum",END)
graph.add_edge("getSub",END)
graph.add_edge("getMul",END)
graph.add_edge("getDiv",END)

workflow=graph.compile()
```

```python
initial_state={"num1":100,"num2": 20}
final_state = workflow.invoke(initial_state)
print(final_state)
```
